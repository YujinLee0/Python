```python
f = open('students.txt', 'r')


list_a = [ ]

stdDict = { }

a = f.read().split("\n")

# split해서 리스트에 넣어주기
for line in a:
    list_a.append(line)
for l in list_a:
    if l == "":
        break
    b = l.split("\t")
    print(b)
    stdDict[b[0]] = [b[1],b[2],b[3]]


stdDict # 학번을 key로 받는 딕셔너리 생성 
```

```python
for i in stdDict:
    for s in stdDict[i]:
        print(s)
```

```python
class GradeSystem:
    
    def __init__(self):
        self.std = stdDict
    
        # average 추가로 저장하기 
        for i in self.std:
            mid = int(self.std[i][1])
            fin = int(self.std[i][2])
            
            avg = (mid + fin) / 2
            self.std[i] = [self.std[i][0],self.std[i][1],self.std[i][2],avg]
            
        # Grade 추가로 저장하기
        for i in self.std:
            if self.std[i][3] >= 90:
                grade = 'A'
            elif self.std[i][3] >= 80:
                grade = 'B'
            elif self.std[i][3] >= 70:
                grade = 'C'
            elif self.std[i][3] >= 60:
                grade = 'D'
            else:
                grade = 'F'
            self.std[i].extend(grade)
    
    
    def show(self):
        """ 전체 학생 정보 출력 """
        print("Student".rjust(15),"Name".rjust(15),"Midterm".rjust(15),"Final".rjust(15),"Average".rjust(15),"Grade".rjust(15))
        print("-"*100)
        
        # 평균점수 기준 내림차순 출력
        self.std = dict(sorted(self.std.items(), reverse =True, key = lambda x: x[1][3]))
        # sorted - 딕셔너리에 items() 메서드를 사용해주면 {"key" : value}의 형태를 [(key, value)]의 형태로 만들어 준다.
        # 따라서 value에 대한 정렬은 lambda를 통해 [1]로 접근한다. 

        
        for s in self.std:
            print(s.rjust(15),end=" ")
            for i in self.std[s]:
                print(str(i).rjust(15), end=" ")
            print()

    
    def search(self, stdNo):
        """ 특정 학생 검색 """
        print("Student              Name              Midterm        Final           Average         Grade ")
        print("-"*10)
        
        print(stdNo, end = "\t")
        for i in self.std[stdNo]:
            print(i,  end="\t"*2)            
    
    def changeScore(self, stdNo, testNo, score):
        """ 점수 수정 """
        # mid 점수가 바뀐 경우
        if testNo == 'mid':
            self.std[stdNo][1] = score
            self.std[stdNo][3] = (int(self.std[stdNo][1]) + int(self.std[stdNo][2])) / 2
            
            if self.std[stdNo][3] >= 90:
                grade = 'A'
            elif self.std[stdNo][3] > 80:
                grade = 'B'
            elif self.std[stdNo][3] > 70:
                grade = 'C'
            elif self.std[stdNo][3] > 60:
                grade = 'D'
            else:
                grade = 'F'   
            
            self.std[stdNo][4] = grade
            print("수정이 완료되었습니다. ")
        
        # final 점수가 바뀐 경우 
        elif testNo == 'final':
            self.std[stdNo][2] = score
            self.std[stdNo][3] = (int(self.std[stdNo][1]) + int(self.std[stdNo][2])) / 2
            
            if self.std[stdNo][3] >= 90:
                grade = 'A'
            elif self.std[stdNo][3] >= 80:
                grade = 'B'
            elif self.std[stdNo][3] >= 70:
                grade = 'C'
            elif self.std[stdNo][3] >= 60:
                grade = 'D'
            else:
                grade = 'F'   
            
            self.std[stdNo][4] = grade      
            print("수정이 완료되었습니다. ")
                
        
        self.std = dict(sorted(self.std.items(), reverse =True, key = lambda x: x[1][3]))
        
            
    def add(self, stdNo, name, mid, fin):
        """ 학생 추가 """
        avg = (mid + fin) / 2
        
        if avg >= 90:
            grade = 'A'
        elif avg >= 80:
            grade = 'B'
        elif avg >= 70:
            grade = 'C'
        elif avg >= 60:
            grade = 'D'
        else:
            grade = 'F'
        
        self.std[stdNo] = [name, mid, fin, avg, grade]
        self.std = dict(sorted(self.std.items(), reverse =True, key = lambda x: x[1][3]))
        print("Student Added")
    
    def searchGrade(self, grade):
        ppl = [ ]
        for s in self.std:
            if self.std[s][4] == grade:
                ppl.append(s)
        
        print("Student              Name              Midterm        Final           Average         Grade ")
        print("-"*100)    

        for i in ppl:
            print(i, end='\t')
            for l in self.std[i]:
                print(l, end='\t'*2)
            print()
    
    
    def remove(self, stdNo):
        del(self.std[stdNo])
        self.std = dict(sorted(self.std.items(), reverse =True, key = lambda x: x[1][3]))
        print("Student Removed")
    
    def quit(self, fname):
        self.std = dict(sorted(self.std.items(), reverse =True, key = lambda x: x[1][3]))        
        fw = open(fname, 'w')
        a = ' '

        for s in self.std:
            a += str(s)
            for i in self.std[s]:
                a += str(i)
                a += "\t"

            a += "n"
            
        fw.write(a)
        fw.close()
        print("Saved")


        
sms = GradeSystem()
```

```python
# 사용자가 시스템을 이용할 수 있는 코드 
while True:

    flag = input("메뉴를 선택하시오. \n Show | Search | ChangeScore | SearchGrade | Add | Remove  | Quit : ")
    flag = flag.lower()
    
    if flag == 'show':
        sms.show()

    elif flag == 'search':
        stdNo = input("학번을 입력하세요:")
        if stdNo in sms.std:
            sms.search(stdNo)
        if stdNo not in sms.std:
            print('NO SUCH PERSON')
    
    elif flag == 'changescore':
        
        # 학번에 대한 확인
        stdNo = input("학번을 입력하세요:")
        if stdNo in sms.std:
            # 시험 종류에 대한 확인
            testNo = input("mid, final 중 하나를 선택하세요.")
            if testNo == "mid" or "final":
                # 점수 값 확인
                score = float(input("시험 점수를 입력하세요: "))
                if score in range(0,101):    
                    sms.changeScore(stdNo, testNo, score)
        
            break
        
        if stdNo not in sms.std:
            print('NO SUCH PERSON')
            break

        
    
    elif flag == 'add':
        stdNo = input("학생의 학번을 입력하세요: ")
        
        if stdNo in sms.std:
            print("ALREADY EXISTS")
            
        if stdNo not in sms.std:
            name = input("이름을 입력하세요: ")
            mid = float(input("중간고사 점수를 입력하세요: "))
            fin = float(input("기말고사 점수를 입력하세요: "))
            
        sms.add(stdNo, name, mid, fin)
     
    elif flag == 'searchgrade':
        grade = input("Grade to search: ").upper()
        gradelist = ['A','B','C','D','F']
        sList = [ ]
        if grade not in gradelist:
            print("정확한 grade를 입력하세요.")
            break
        if grade in gradelist:
            for s in sms.std:
                if grade == sms.std[s][4]:
                    sList.append(s)
            
            if sList == []:
                print("No RESULTS")
            if len(sList) > 0:
                sms.searchGrade(grade)
                
    elif flag == 'remove':
        stdNo = input("학생의 학번을 입력하세요: ")
        print(sms.std.keys())
        if stdNo in sms.std.keys():
            sms.remove(stdNo)
        elif stdNo not in sms.std.keys():
            print("NO SEARCH PERRSON")
        elif len(sms.std) == 0:
            print("List is empty")
    
    elif flag == 'quit':
        ans = input("Save data? [yes/no]")
        if ans == 'yes':
            fname = input("file name: ")
            sms.quit(fname)
        if ans == 'no':
            break
        
        
    else:
        continue
        
    
```
